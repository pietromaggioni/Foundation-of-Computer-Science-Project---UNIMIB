1. Create a single dataframe with the concatenation of all input csv files, adding a column called country

```python
file_path = '/Users/pietro/Documents/Project/trendingYT'
```

```python
dfs = []

for filename in os.listdir(file_path):
    if filename.endswith(".csv.zst"):
        country = filename[:2]   
        filepath = os.path.join(file_path, filename)

        df_temp = pd.read_csv(filepath, compression = 'zstd', encoding='utf-8', encoding_errors = 'ignore')
        df_temp["country"] = country
        print(f'Processed {filename}')
        dfs.append(df_temp)

final_df = pd.concat(dfs, ignore_index=True)
```

```python
final_df.sample(5)
```

2. Extract all videos that have no tag.

```python
#sostiuisco con NaN i [none] per tutto il dataframe e formatto la data 
final_df.replace('[none]', np.nan, inplace=True)
final_df['tags'] = final_df['tags'].replace('', np.nan)
final_df['trending_date'] = pd.to_datetime(final_df['trending_date'], format='%y.%d.%m')
df_noTAG = final_df[final_df['tags'].isna()]
```

```python
print(f'There are {df_noTAG.shape[0]} videos that have no tags')
```

3. For each channel, determine the total number of views

```python
grouped = final_df.groupby('channel_title')
print(grouped.agg({
    'views': 'sum'
}).sort_values(by='views', ascending = False))
```

4. Save all rows with disabled comments and disabled ratings, or that have video_error_or_removed in a new dataframe called excluded, and remove those rows from the original dataframe.

```python

excluded = final_df[((final_df['ratings_disabled'] == True) & (final_df['comments_disabled'] == True)) | 
                     (final_df['video_error_or_removed'] == True)]
```

```python
print(f'{excluded.shape[0]} have been excluded from the data set')
final_df.shape[0]
```

```python
excluded.head(5)
```

```python
print(f'Starting number of rows: {final_df.shape[0]}')
final_df = final_df.drop(excluded.index, errors='ignore')
final_df = final_df.reset_index(drop=True)
print(f'Final number of rows: {final_df.shape[0]}')
final_df.head(5)
```

5. Add a like_ratio column storing the ratio between the number of likes and of dislikes

```python
final_df = final_df.assign(LikesRatio = lambda x: np.where(x['dislikes'] == 0, x['likes'], x['likes'] / x['dislikes']))
```

```python
final_df.head()
```

6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
final_df['publish_time'] = pd.to_datetime(final_df['publish_time'])
final_df['start_interval'] = final_df['publish_time'].dt.floor('10min')
final_df['end_interval'] = final_df['start_interval'] + pd.Timedelta(minutes=10)
final_df['publish_interval'] = final_df['start_interval'].dt.strftime('%H:%M') + '-' + final_df['end_interval'].dt.strftime('%H:%M')
final_df.drop(columns=['start_interval', 'end_interval'], inplace=True)
```

```python
final_df.head(5)
```

7. For each interval, determine the number of videos, average number of likes and of dislikes.

```python
analytics = final_df.groupby('publish_interval')
analytics.agg({
    'video_id': 'count', 
    'likes': 'mean', 
    'dislikes': 'mean', 
    
})
```

8. For each tag, determine the number of videos

```python

all_tags = final_df['tags'].str.replace('"', '', regex=False).str.split('|').explode()
all_tags = all_tags[all_tags != '']
tag_counts = all_tags.value_counts().reset_index()
tag_counts.columns = ['tag', 'num_videos']
print(tag_counts)
```

9. Find the tags with the largest number of videos

```python
maxNumber = tag_counts['num_videos'].max()
max_tag = tag_counts[tag_counts['num_videos'] == maxNumber]['tag'][0]
print(f'the tag with the largest numer of videos is: {max_tag}')
```

 10. For each (tag, country) pair, compute average ratio likes/dislikes

```python
#creation of a temporary df to avoid memory problems --> expoding the tags column will create a large number of additional rows 
temp_df = final_df[['tags', 'likes', 'dislikes', 'country', 'LikesRatio']].copy()
temp_df.dropna(subset=['tags'], inplace=True)
temp_df['tag'] = temp_df['tags'].str.split('|')
temp_df = temp_df.explode('tag')
temp_df = temp_df[temp_df['tag'] != '']   #before adding this line i had an empty tag per contry in the groupby --> possible presence of '||'
group = temp_df.groupby(['tag', 'country'])
aggr = group.agg({
    'LikesRatio': 'mean'
})
print('Sorted')
aggr.sort_values(by='LikesRatio', ascending = False)
```

14. Read all json files with the video categories

```python
dfs = []

for filename in os.listdir(file_path):
    if filename.endswith(".json"):
        country = filename[:2] 
        filepath = os.path.join(file_path, filename)
        with open(filepath) as j:
            data = json.load(j)
    
            items_data = data['items']
            df_temp = pd.json_normalize(items_data)
            df_temp["country"] = country
            df_temp['kind'] = data['kind']
            df_temp['etag'] = data['etag']
            print(f'file {filename} processsed')
        dfs.append(df_temp)

categories_df = pd.concat(dfs, ignore_index=True)
categories_df = categories_df.rename(columns = {'id': 'category_id'})
categories_df.head(5)
```

15. For each country, determine how many videos have a category that is not assignable.

```python

categories_df['category_id'] = categories_df['category_id'].astype('int64')    #it had object type before so the merging was failing 
final_df = pd.merge(final_df, categories_df, on=['category_id', 'country'])
```
